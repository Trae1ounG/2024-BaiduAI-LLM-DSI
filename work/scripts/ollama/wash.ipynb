```python
import json
import re
from tqdm import tqdm
from collections import defaultdict
import pandas as pd
input_file = 'train_indexing_qwen_merge.json'
```

```python
with open(input_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()
    data = [json.loads(line) for line in lines]
    

```

```python
pd_data = pd.DataFrame(data
pd_data = pd_data.drop_duplicates(subset=['tgt'])
pd_data = pd_data.dropna()
```

```python

# 按照 tgt 列排序
pd_data = pd_data.sort_values(by='tgt', key=lambda x: x.astype(int))

```

```python

# 定义符号映射
symbol_map = {
    '；': ';',
    '，': ',',
    '。': '.',
    '？': '?',
    '！': '!',
    '：': ':',
    '“': '"',
    '”': '"',
    '‘': "'",
    '’': "'",
    '（': '(',
    '）': ')',
    '【': '[',
    '】': ']',
    '《': '<',
    '》': '>'
}

# 替换中文符号为英文符号
def replace_symbols(text):
    for zh_symbol, en_symbol in symbol_map.items():
        text = text.replace(zh_symbol, en_symbol)
    return text

# 应用替换函数到 DataFrame 的指定列
pd_data['src'] = pd_data['src'].apply(replace_symbols)

# 定义删除网址的函数
def remove_urls(text):
    return re.sub(r'http[s]?://\S+', '', text)

# 使用apply函数删除网址
pd_data['src'] = pd_data['src'].apply(remove_urls)


```

```python
pd_data['src'] = pd_data['src'].str.replace(' ', ';')
pd_data['src'] = pd_data['src'].str.replace('、', ';')
pd_data['src'] = pd_data['src'].str.replace(',', ';')
pd_data['src'] = pd_data['src'].str.lstrip(':')
```

```python
import re

# 合并两个列表并创建正则表达式模式
all_search_terms = search_terms + search_terms2+ search_terms3
pattern = '|'.join(map(re.escape, all_search_terms))

# 删除搜索词和符号
pd_data['src'] = pd_data['src'].apply(lambda x: re.sub(pattern, '', x))
pd_data['src'] = pd_data['src'].str.lstrip(':')
print("\nDataFrame after removing search terms and symbols:")
pd_data = pd_data.reset_index(drop=True)
```
